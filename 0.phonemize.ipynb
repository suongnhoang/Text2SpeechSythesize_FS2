```python
from pathlib import Path

txt_files = sorted(Path("./Data/Infore/Infore").glob("*.txt"))
words = []
for txt_file in txt_files:
    wav_file = txt_file.with_suffix(".wav")
    if not wav_file.exists():
        continue
    line = open(txt_file, "r", encoding="utf-8").read()
    words.extend(line.strip().lower().split())

words = list(sorted(set(words)))

with open("words.txt", "w", encoding="utf-8") as f:
    for word in words:
        f.write(word)
        f.write("\n")
```

```python

englist = "ôtô,kitô,album,asean,bar,calo,casino,coca,euro,gara,gas,guitar,ion,karaoke,kelvin,laser,mangan,mark,martin,mini,ok,radio,real,robot,rock,sex,shop,sol,stress,studio,taxi,test,toilet,ts,video,virus,visa,yoga".split(",")

black_list = (
    englist
    + ["q", "adn", "h", "b", "k", "cs", "l", "ad", "hiv"]
    + ["v", "d", "c", "p", "x"]
    + ["đt", "đicr", "đigiúp"]
    + ["khmer", "ii", "s", "tr", "xhcn", "hlv"]
    + ["qúy", "hlv" "ts", "usd", "ph"]
    + ["s", "kg", "km", "g", "tr", "đ", "ak", "d", "m", "n"]
)

N_onsets = { u'b' : u'b', u't' : u't', u'th' : u'tʰ', u'đ' : u'd', u'ch' : u'c', 
            u'kh' : u'x', u'g' : u'ɣ', u'l' : u'l', u'm' : u'm', u'n': u'n', 
            u'ngh': u'ŋ', u'nh' : u'ɲ', u'ng' : u'ŋ', u'ph' : u'f', u'v' : u'v', 
            u'x' : u's', u'd' : u'z', u'h' : u'h', u'p' : u'p', u'qu' : u'kw',
            u'gi' : u'z', u'tr' : u'c', u'k' : u'k', u'c' : u'k', u'gh' : u'ɣ', 
            u'r' : u'z', u's' : u's', u'gi': u'z'}

pad = list(set(N_onsets.values()))
Special=['tʰ','tʰw','p','pw','t','tw','ʈ','ʈw','c','cw','k','kw','k','kw','kw',"'",'b','bw',
        'd','dw','f','fw','s','sw','ʂ', 'ʂw', 'x', 'xw', 'h', 'hw', 'v', 'vw', 'z', 'zw', 
        'j', 'jw', 'ʐ', 'ʐw', 'ɣ', 'ɣw', 'ɣ', 'ɣw', 'm', 'mw', 'n', 'nw', 'ɲ', 'ɲw', 
        'ŋ', 'ŋw', 'ŋ', 'ŋw', 'l', 'lw',"'"]
pad += Special

def g2p(w):
    w = w.strip()
    if w not in black_list:
        a = vi2IPA_split(w,"/")\
                .replace("/./","")\
                .strip()\
                .split("/")[1:-1]
        
        a = "|".join(a).replace("_", " ")
        ps = a.split("| |")
        res = []
        for a in ps:
            a = a.split("|")
            tone = a[-1] if a[-1].isdigit() else None
            
            if tone is not None:
                if int(tone) in [0,1,2,3,4,5,6]:
                    if a[0] in pad:
                        for i in range(1,len(a)):
                            a[i] += tone
                    else:
                        for i in range(0,len(a)):
                            a[i] += tone
                        a = ["silp"] + a
                    a = a[:-1]
            res.extend(a)
        return res
    return []
```

```python
ws = open("./words.txt").readlines()
f = open("./lexicon/infore-lexicon.txt", "w")
for w in sorted(set(ws)):
    w = w.strip()
    # this is a hack to match phoneme set in the vietTTS repo
    p = g2p(w)
    p = " ".join(p)
    if w in black_list or p == "":
        continue
    else:
        f.write(f"{w}\t{p}\n")
f.close()
```
